```python
import numpy as np
import pandas as pd
import glob
import cv2 as cv
import os

## 데이터 증식

step = 1
size = 50
angleRange = 90
angleRange //= 2
angleRange //= step

path = 'D:\Project 2'

os.chdir(path)

path += '\svgtopng\\'

emotions = tuple('angry,happy,neutral,sad,surprised'.split(','))

for emotion in emotions:
    
    newPath = path + emotion
    files = glob.glob(newPath + '\*.png')
    removeFiles = [i for i in files if len(i.split('_')) != 1]
    files = [i for i in files if len(i.split('_')) == 1]
    
    for f in removeFiles:
        os.remove(f)
        
    #continue
    for f in files:
        img = cv.imread(f,cv.IMREAD_UNCHANGED)
        
        variant = img.shape[2]
        
        if img.shape[2] != 4:
            trial = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
        else:
            row = len(img)
            col = len(img[0])
            trial = img[:,:,3]
            #trial = [[255]*col]*row - trial
            trial = np.asarray(trial)
        #trial = [[255]*col]*row - img[:,:,-1]
        
        img = cv.resize(src=trial, dsize=(size,size))
        if variant == 4:
            img = ~img
        
        title, ext = os.path.splitext(f)
        cv.imwrite(title + '_resized' + ext,img)
        
        
        center = (size//2,size//2)
        scale = 1.0
        
        for i in range(-angleRange,angleRange+1,step):
            if i == 0:
                continue
            mtx = cv.getRotationMatrix2D(center, i, scale)
            alpha = i * step# if i >= 0 else (360 - step*i)
            img2 = cv.warpAffine(img, mtx, (size,size))
            cv.imwrite(title + "_resized_%d" % (i*step) + ext, img2)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import glob
import cv2 as cv
import os

## 데이터 증식

#step = 1
maxData = 500

path = 'D:\Project 2'

os.chdir(path)

path += '\svgtopng\\'

xFiles = []

emotions = tuple('angry,happy,neutral,sad,surprised'.split(','))
y = []
for i in range(len(emotions)):
    
    newPath = path + emotions[i]
    files = glob.glob(newPath + '\*.png')
    files = [j for j in files if len(j.split('_')) != 1]
    if len(files) > maxData:
        files = files[:maxData]
    xFiles.extend(files)
    y.extend([emotions[i]]*len(files))
    
y = pd.Series(y)
y = pd.get_dummies(y,prefix='y',prefix_sep="_").astype('int32')

y = np.array(y).astype('int32')
X = []
i = 0
for f in xFiles:
    
    img = cv.imread(f, cv.IMREAD_GRAYSCALE)
    #img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
    #img = img.resize((49,49))
    img = np.asarray(img)
    img = img.astype('float32') / 255
    #img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
     #img = img.convert('RGBA')
    #img = img.convert("A")
    #img = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
    X.append(img)

#import tensorflow.keras
import numpy as np
#from tensorflow.keras.utils import to_categorical

# 입력과 출력 지정하기 
nb_classes = len(emotions) # 레이블 수
X = np.array(X)
X = X.reshape(-1, size,size,1)
```

```python
X = np.array(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,
                                                   test_size=0.2,
                                                   random_state=777,
                                                   stratify=y)

X_train, X_val, y_train, y_val = train_test_split(X_train, y_train,
                                                 test_size=0.2,
                                                 random_state=777,
                                                 stratify=y_train)
```

```python
import tensorflow.keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Flatten
from tensorflow.keras.layers import Conv2D, MaxPooling2D
from tensorflow.keras.optimizers import Adam

# CNN 모델 정의하기
model = Sequential([
    Conv2D(32, kernel_size=(3,3), activation='relu',input_shape=(size,size, 1)),
    Conv2D(32, kernel_size=(3,3), activation='relu'),
    MaxPooling2D(pool_size=(2,2)),
    Dropout(0.25),
        
    Conv2D(64, kernel_size=(3,3), activation='relu'),
    Conv2D(64, kernel_size=(3,3), activation='relu'),
    MaxPooling2D(pool_size=(2,2)),
    Dropout(0.25),
    
    Conv2D(128, kernel_size=(3,3), activation='relu'),
    Conv2D(128, kernel_size=(3,3), activation='relu'),
    MaxPooling2D(pool_size=(2,2)),
    Dropout(0.25),
        
    Flatten(),
    Dense(512, activation='relu'),
    Dropout(0.5),
    Dense(nb_classes,activation='softmax')
        
])
```

```python
model.summary()
```

```python
model.compile(loss='categorical_crossentropy',
             optimizer=Adam(1e-4),
             metrics=['accuracy'])
```

```python
hist = model.fit(X_train,y_train, 
         batch_size=1,
         epochs=30,
         verbose=1,
         validation_data=(X_val, y_val))
```

```python
# 모델 평가하기 
score = model.evaluate(X_test, y_test, verbose=1)
print('정답률=', score[1], 'loss=', score[0])
```

```python
finalModel = Sequential([
    Conv2D(32, kernel_size=(3,3), activation='relu',input_shape=(size,size, 1)),
    MaxPooling2D(pool_size=(2,2)),
    Dropout(0.25),
        
    Conv2D(64, kernel_size=(3,3), activation='relu'),
    MaxPooling2D(pool_size=(2,2)),
    Dropout(0.25),
        
    Flatten(),
    Dense(512, activation='relu'),
    Dropout(0.5),
    Dense(nb_classes,activation='softmax')
        
])
finalModel.compile(loss='categorical_crossentropy',
                     optimizer=Adam(1e-4),
                     metrics=['accuracy'])
histF = finalModel.fit(X,y, 
         batch_size=16,
         epochs=10,
         verbose=1)
```
